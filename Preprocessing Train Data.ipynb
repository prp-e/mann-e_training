```python
import pandas as pd
import cv2
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.tail()
```

```python
raw_images = 'raw_images'
train_data = 'train_data'
encoder_data = 'encoder_data'
```

```python
for row in df.iterrows():
    file_name = row[1]['file']
    file_name = file_name.split('.png')[0]
    file_name = f'{file_name}.txt'
    
    prompt = row[1]['prompt']
    
    f = open(f'{encoder_data}/mstyle-{file_name}', 'w')
    f.write(prompt)
    f.close()
```

```python
from matplotlib import pyplot as plt
```

```python
import os
```

```python
names = os.listdir(raw_images)
```

```python
images = []

for name in names:
    name = f'{raw_images}/{name}'
    image = cv2.imread(name)
    
    images.append(image)
```

```python
len(images)
```

```python
plt.imshow(cv2.cvtColor(images[185], cv2.COLOR_BGR2RGB))
```

```python
resized_images = [] 

for image in images:
    image = cv2.resize(image, [512, 512])
    resized_images.append(image)
```

```python
plt.imshow(cv2.cvtColor(resized_images[185], cv2.COLOR_BGR2RGB))
```

```python
for name, image in zip(names, resized_images):
    cv2.imwrite(f'{encoder_data}/mstyle-{name}', image)
```
